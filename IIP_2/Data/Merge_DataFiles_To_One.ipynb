## Leerstehende Wohnungen Excel files in eine CSV Datei umwandeln

```python
import os
import pandas as pd
import re

# Define the directory where the Excel files are located
input_dir = './Raw/Leerwohnungsziffer/'

# Get a list of all the Excel files in the input directory
files = os.listdir(input_dir)
excel_files = [f for f in files if f.endswith('.xlsx')]

# Load the first Excel file into a Pandas DataFrame
df_merged = pd.read_excel(os.path.join(input_dir, excel_files[0]), index_col=None, header=None, skiprows=7, skipfooter=5)
# change colum names
df_merged.columns = ['Gemeinde', 'PLZ', 'Leerwohnungen', 'Einfamilien', 'Neubauten', 'Vermietbar', '1_Zimmer', '2_Zimmer', '3_Zimmer', '4_Zimmer', '5_Zimmer', '6_Zimmer_und_mehr', 'Leerwohnungen_in_Prozent']
# add year
grp = re.search(r'(?P<jear>\d{4})_001\.xlsx$', excel_files[0], re.MULTILINE)
df_merged['Jahr'] = int(grp.group('jear'))

# Loop through each remaining Excel file and merge it with the existing DataFrame
for file in excel_files[1:]:
    df = pd.read_excel(os.path.join(input_dir, file), index_col=None, header=None, skiprows=7, skipfooter=5)
    df.columns = ['Gemeinde', 'PLZ', 'Leerwohnungen', 'Einfamilien', 'Neubauten', 'Vermietbar', '1_Zimmer', '2_Zimmer', '3_Zimmer', '4_Zimmer', '5_Zimmer', '6_Zimmer_und_mehr', 'Leerwohnungen_in_Prozent']
    # add year
    grp = re.search(r'(?P<jear>\d{4})_001\.xlsx$', file, re.MULTILINE)
    df['Jahr'] = int(grp.group('jear'))
    df_merged = pd.concat([df_merged, df])

# Reset the index   
df_merged.reset_index(drop=True, inplace=True)

# replace - and ... with 0
df_merged = df_merged.replace('-', 0)
df_merged = df_merged.replace('...', 0)

# Save the merged DataFrame as a new Excel file
output_path = './Preparation/merged_leerwohnungen.csv'
df_merged.to_csv(output_path, index=False)
```

## Wohnungsflächen Excel in ein CSV umwandeln

```python
import os
import pandas as pd
import re

# Define the directory where the Excel files are located
input_dir = './Raw/Wohnungsflaeche/'

# Get a list of all the Excel files in the input directory
files = os.listdir(input_dir)
excel_files = [f for f in files if f.endswith('.xlsx')]
# Load the first Excel file into a Pandas DataFrame
df_merged = pd.read_excel(os.path.join(input_dir, excel_files[0]), index_col=None, header=None, skiprows=7, skipfooter=3)
if len(df_merged.columns) == 10:
    # add "ohne angaben"
    df_merged['ohne_Angaben'] = 0
# change colum names
df_merged.columns = ['Bauperiode', 'Wohnungen_Total', 'Davon_Einfamilienhaeuser', '1or2_Zimmer', '3or4_Zimmer', '5_Zimmer_und_mehr', '0to49m', '50to99m', '100to149m', '150m_und_mehr', 'ohne_Angaben']
# add "ohne angaben"
df_merged['ohne_Angaben'] = 0
# add year
grp = re.search(r'(?P<jear>\d{4})_001\.xlsx$', excel_files[0], re.MULTILINE)
df_merged['ExcelDate'] = int(grp.group('jear'))

# Loop through each remaining Excel file and merge it with the existing DataFrame
for file in excel_files[1:]:
    df = pd.read_excel(os.path.join(input_dir, file), index_col=None, header=None, skiprows=7, skipfooter=3)
    if len(df.columns) == 10:
        # add "ohne angaben"
        df['ohne_Angaben'] = 0
    df.columns = ['Bauperiode', 'Wohnungen_Total', 'Davon_Einfamilienhaeuser', '1or2_Zimmer', '3or4_Zimmer', '5_Zimmer_und_mehr', '0to49m', '50to99m', '100to149m', '150m_und_mehr', 'ohne_Angaben']
    # add year
    grp = re.search(r'(?P<jear>\d{4})_001\.xlsx$', file, re.MULTILINE)
    df['ExcelDate'] = int(grp.group('jear'))
    df_merged = pd.concat([df_merged, df], ignore_index=True)

# Reset the index   
df_merged.reset_index(drop=True, inplace=True)

# replace - and ... with 0
df_merged = df_merged.replace('-', 0)
df_merged = df_merged.replace('...', 0)

# Save the merged DataFrame as a new Excel file
output_path = './Preparation/wohnungsflaechen.csv'
df_merged.to_csv(output_path, index=False)
```
